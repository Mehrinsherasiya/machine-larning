```python
#Image Sharpening
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('nature.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Creating the kerenel
k1=np.array([[-1,-1,-1],[-1,-1,-1],[-1,-1,-1]])
k2=np.array([[1,1,1],[1,-7,1],[1,1,1]])
k3=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
k4=np.array([[-1,-1,-1,-1,-1],[-1,2,2,2,-1],[-1,2,8,2,-1],[-1,-1,-1,-1,-1]])/8


#Applyin the kerenel
img1=cv2.filter2D(img,-1,k1)
img2=cv2.filter2D(img,-1,k2)
img3=cv2.filter2D(img,-1,k3)
img4=cv2.filter2D(img,-1,k4)


plt.figure(figsize=(10,6))
plt.subplot(3,2,1),plt.imshow(img),plt.title('Source image'),plt.axis('off')
plt.subplot(3,2,2),plt.imshow(img1),plt.title('k1'),plt.axis('off')
plt.subplot(3,2,3),plt.imshow(img2),plt.title('k2'),plt.axis('off')
plt.subplot(3,2,4),plt.imshow(img3),plt.title('k3'),plt.axis('off')
plt.subplot(3,2,5),plt.imshow(img4),plt.title('k4'),plt.axis('off')
```

```python
#Image Sharpening
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('hd.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

gauss=cv2.GaussianBlur(img,(5,5),0)
sharped_img=cv2.addWeighted(img,1.5,gauss,-0.5,0)

plt.subplot(1,2,1),plt.imshow(img),plt.title('Source image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(sharped_img),plt.title('sharped_img'),plt.axis('off')
```
