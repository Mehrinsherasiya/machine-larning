```python
#Applying advance image filters
#1).Bilateral (for advanced smoothing effect)
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('rose.jpg')
bilateral=cv2.bilateralFilter(img,d=9,sigmaColor=95,sigmaSpace=95)
plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('Source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(bilateral,cv2.COLOR_BGR2RGB)),plt.title('with smoothing effect'),plt.axis('off')
```

```python
#Applying advance image filters
#2).Unsharp (for advanced smoothing effect)
import cv2
import matplotlib.pyplot as plt
#Taking the image
img=cv2.imread('aerial.jpg')
#Blur the image
gauss=cv2.GaussianBlur(img,(9,9),0)
#Unsharp masking
unsharp=cv2.addWeighted(img,1.5,gauss,-0.5,0)
plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('Source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(unsharp,cv2.COLOR_BGR2RGB)),plt.title('with sharpening effect'),plt.axis('off')
```

```python
#Applying advance image filters
#1).Scharr (for advanced smoothing effect)
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('zebra.jpg',0) #0 gray scale
#Applying Scharr filters
scharrx=cv2.Scharr(img,cv2.CV_64F,1,0)#vertical edges
scharry=cv2.Scharr(img,cv2.CV_64F,0,1)#horizontal edges
#Converting to unit8 for display
scharrx=cv2.convertScaleAbs(scharrx)
scharry=cv2.convertScaleAbs(scharry)
#Combine both directions
scharr_combine=cv2.addWeighted(scharrx,0.10,scharry,0.2,0)
#Plotting the image
plt.subplot(1,3,1),plt.imshow(scharrx,cmap='gray'),plt.title('Scharr-X'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(scharry,cmap='gray'),plt.title('Scharr-Y'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(scharr_combine,cmap='gray'),plt.title('Scharr'),plt.axis('off')
```
